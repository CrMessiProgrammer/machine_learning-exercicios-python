## Construção de gráfico a partir de x e y

```
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y)
plt.plot(x,y, 'ro')   #'ro' significa r(red) e o(formato das marcações no gráfico)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico x por y')
plt.grid()
```

## Construção de gráfico a partir de funções

```
import numpy as np
```

```
def f(t):
  return np.exp(-t)*np.cos(2*np.pi*t)

  t1=np.arange(0.0,5.0,0.1)
  t2=np.arange(0.0,5.0,0.02)

  plt.figure()
  plt.subplot(211)
  plt.grid()
  plt.plot(t1,f(t1),'bo', t2,f(t2),'k')
```

```
t3=np.arange(0.0,4.0,0.01)
plt.subplot(212)
plt.plot(t3,np.exp(-t3)*np.cos(2*np.pi*t3),'k')
plt.show()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfico x por y')
```
